#### 데이터 전처리

```python
df = df[
    (df['country_iso3_code'] == 'KOR') |
    (df['country_iso3_code'] == 'CHN') |
    (df['country_iso3_code'] == 'USA') |
    (df['country_iso3_code'] == 'JPN') |
    (df['country_iso3_code'] == 'VNM')]

pd.DataFrame(df)
```

```python
df = df.drop(columns='country_id')

pd.DataFrame(df)
```

```python
df = df.drop(columns='product_id')

pd.DataFrame(df)
```

```python
def resetNum(code):
    if 1 <= code <= 5:
        return 1
    elif 6 <= code <= 14:
        return 2
    elif code == 15:
        return 3
    elif 16 <= code <= 24:
        return 4
    elif 25 <= code <= 27:
        return 5
    elif 28 <= code <= 38:
        return 6
    elif 39 <= code <= 40:
        return 7
    elif 41 <= code <= 43:
        return 8
    elif 44 <= code <= 46:
        return 9
    elif 47 <= code <= 49:
        return 10
    elif 50 <= code <= 63:
        return 11
    elif 64 <= code <= 67:
        return 12
    elif 68 <= code <= 70:
        return 13
    elif code == 71:
        return 14
    elif 72 <= code <= 83:
        return 15
    elif 84 <= code <= 85:
        return 16
    elif 86 <= code <= 89:
        return 17
    elif 90 <= code <= 92:
        return 18
    elif code == 93:
        return 19
    elif 94 <= code <= 96:
        return 20
    elif code == 97:
        return 21
    else:
        return 22

df['product_hs12_code'] = df['product_hs12_code'].apply(resetNum)
pd.DataFrame(df)
```

## [데이터 분석]

#### 1. 한국의 수출입 구조와 변화 분석

```python
kor = df[df['country_iso3_code'] == "KOR"]
pd.DataFrame(kor)
```

한국의 수출입 변화 추이

```python
kor_p = kor.groupby('year')[['export_value','import_value']].sum()
kor_p
```

```python
kor_p.plot(kind='line')
```

한국의 무역 순이익

```python
kor_net_profit = kor_p['export_value'] - kor_p['import_value']
pd.DataFrame(kor_net_profit)
```

```python
kor_net_profit.plot(kind='line')
```

한국의 제품군별 수출액 변화

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

kor_grouped_ex = kor.groupby(['year', 'product_hs12_code'])['export_value'].sum().reset_index()

plt.figure(figsize=(14, 8))

sns.lineplot(data=kor_grouped_ex, x='year', y='export_value', hue='product_hs12_code', marker='o', palette='tab20', legend=False)

# 각 선 끝에 라벨 추가
for code in kor_grouped_ex['product_hs12_code'].unique():
    data = kor_grouped_ex[kor_grouped_ex['product_hs12_code'] == code]
    last_point = data.iloc[-1]
    plt.text(last_point['year'], last_point['export_value'], str(code), fontsize=10, fontweight='bold', va='center', ha='left')

plt.title('Graph of Changes in Korea Exports by Product Line')
plt.xlabel('Year')
plt.ylabel('Export Value')
plt.grid()
plt.show()
```

한국의 제품군별 수입액 변화

```python
kor_grouped_im = kor.groupby(['year', 'product_hs12_code'])['import_value'].sum().reset_index()

plt.figure(figsize=(14, 8))

sns.lineplot(data=kor_grouped_im, x='year', y='import_value', hue='product_hs12_code', marker='o', palette='tab20', legend=False)

for code in kor_grouped_im['product_hs12_code'].unique():
    data = kor_grouped_im[kor_grouped_im['product_hs12_code'] == code]
    last_point = data.iloc[-1]
    plt.text(last_point['year'], last_point['import_value'], str(code), fontsize=10, fontweight='bold', va='center', ha='left')

plt.title('Graph of Changes in Korea Imports by Product Line')
plt.xlabel('Year')
plt.ylabel('Import Value')
plt.grid()
plt.show()
```

#### 2. 한국 주요 교역국과의 무역수지 비교

```python
chn = df[df['country_iso3_code'] == "CHN"]
usa = df[df['country_iso3_code'] == "USA"]
jpn = df[df['country_iso3_code'] == "JPN"]
vnm = df[df['country_iso3_code'] == "VNM"]
```

```python
chn_p = chn.groupby('year')[['export_value','import_value']].sum()
usa_p = usa.groupby('year')[['export_value','import_value']].sum()
jpn_p = jpn.groupby('year')[['export_value','import_value']].sum()
vnm_p = vnm.groupby('year')[['export_value','import_value']].sum()
```

```python
area1 = plt.subplot(2,2,1)
area2 = plt.subplot(2,2,2)
area3 = plt.subplot(2,2,3)
area4 = plt.subplot(2,2,4)

chn_p.plot(kind='line',ax=area1)
usa_p.plot(kind='line',ax=area2)
jpn_p.plot(kind='line',ax=area3)
vnm_p.plot(kind='line',ax=area4)

area1.set_title('China')
area2.set_title('USA')
area3.set_title('Japan')
area4.set_title('Vietnam')

plt.tight_layout() # 그래프 끼리 겹쳐 보이지 않게 조정(구글링으로 사용/ 밑의 그래프에서 계속 사용)
plt.show()
```

각 국의 순이익 비교

```python
chn_net_profit = chn_p['export_value'] - chn_p['import_value']
usa_net_profit = usa_p['export_value'] - usa_p['import_value']
jpn_net_profit = jpn_p['export_value'] - jpn_p['import_value']
vnm_net_profit = vnm_p['export_value'] - vnm_p['import_value']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.plot(chn_net_profit.index, chn_net_profit.values, label='China', color='red', marker='o')
plt.plot(usa_net_profit.index, usa_net_profit.values, label='USA', color='blue', marker='s')
plt.plot(jpn_net_profit.index, jpn_net_profit.values, label='Japan', color='green', marker='^')
plt.plot(vnm_net_profit.index, vnm_net_profit.values, label='Vietnam', color='purple', marker='d')
plt.plot(kor_net_profit.index, kor_net_profit.values, label='Korea', color='orange', marker='x')

plt.title('Net Trade Profit (Export - Import) by Country')
plt.xlabel('Year')
plt.ylabel('Net Profit')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

각 국 수출액 top5

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

countries = ['USA', 'CHN', 'JPN', 'VNM']

fig = plt.figure(figsize=(24, 6))

area1 = fig.add_subplot(1, 4, 1)
area2 = fig.add_subplot(1, 4, 2)
area3 = fig.add_subplot(1, 4, 3)
area4 = fig.add_subplot(1, 4, 4)
areas = [area1, area2, area3, area4]

for idx, country in enumerate(countries):

    df_country = df[df['country_iso3_code'] == country]
    
    total_exports = df_country.groupby('product_hs12_code')['export_value'].sum().reset_index()
    
    top5_codes = total_exports.sort_values(by='export_value', ascending=False).head(5)['product_hs12_code'].tolist()
    
    df_top5 = df_country[df_country['product_hs12_code'].apply(lambda x: x in top5_codes)]
    
    grouped_ex = df_top5.groupby(['year', 'product_hs12_code'])['export_value'].sum().reset_index()
    
    sns.lineplot(x='year', y='export_value', hue='product_hs12_code', marker='o', palette='muted', legend=False, errorbar=None, data=grouped_ex, ax=areas[idx])
    
    for code in grouped_ex['product_hs12_code'].unique():
        data = grouped_ex[grouped_ex['product_hs12_code'] == code]
        last_point = data.iloc[-1]
        areas[idx].text(last_point['year'], last_point['export_value'], str(code), fontsize=8, fontweight='bold', va='center', ha='left')
    
    areas[idx].set_title(f'{country} Exports (Top 5)')
    areas[idx].set_xlabel('Year')
    areas[idx].set_ylabel('Export Value')
    areas[idx].grid()

fig.suptitle('Exports by Top 5 Product Lines Across Countries (2012-2023)', fontweight='bold')
fig.tight_layout()
plt.show()
```

각 국 수입액 top5

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

countries = ['USA', 'CHN', 'JPN', 'VNM']

fig = plt.figure(figsize=(24, 6))

area1 = fig.add_subplot(1, 4, 1)
area2 = fig.add_subplot(1, 4, 2)
area3 = fig.add_subplot(1, 4, 3)
area4 = fig.add_subplot(1, 4, 4)
areas = [area1, area2, area3, area4]

for idx, country in enumerate(countries):
    
    df_country = df[df['country_iso3_code'] == country]

    total_imports = df_country.groupby('product_hs12_code')['import_value'].sum().reset_index()

    top5_codes = total_imports.sort_values(by='import_value', ascending=False).head(5)['product_hs12_code'].tolist()

    df_top5 = df_country[df_country['product_hs12_code'].apply(lambda x: x in top5_codes)]

    grouped_im = df_top5.groupby(['year', 'product_hs12_code'])['import_value'].sum().reset_index()
    
    sns.lineplot(x='year', y='import_value', hue='product_hs12_code', marker='o', palette='dark', legend=False, errorbar=None, data=grouped_im, ax=areas[idx])
    
    for code in grouped_im['product_hs12_code'].unique():
        data = grouped_im[grouped_im['product_hs12_code'] == code]
        last_point = data.iloc[-1]
        areas[idx].text(last_point['year'], last_point['import_value'], str(code), fontsize=8, fontweight='bold', va='center', ha='left')
    
    areas[idx].set_title(f'{country} Imports (Top 5)')
    areas[idx].set_xlabel('Year')
    areas[idx].set_ylabel('Import Value')
    areas[idx].grid()

fig.suptitle('Imports by Top 5 Product Lines Across Countries (2012-2023)', fontweight='bold')
fig.tight_layout()
plt.show()
```

글로벌 시장 점유율 비교하기
한국의 주요 수출품 16,17,6 번을 기준으로

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

countries = ["CHN", "USA", "JPN", "VNM", "KOR"]
hs_codes = [6, 16, 17]
results = []

for country in countries:
    for hs_code in hs_codes:
        cal = (df["country_iso3_code"] == country) & (df["product_hs12_code"] == hs_code)
        total_share = df[cal]["global_market_share"].sum()
        results.append({"country": country, "hs_code": hs_code, "global_market_share": total_share})

result_df = pd.DataFrame(results)

# 그래프 그리기
plt.figure(figsize=(10, 6))
sns.barplot(data=result_df, x="country", y="global_market_share", hue="hs_code")
plt.title("Global Market Share by Country and HS Code (6, 16, 17)")
plt.xlabel("Country")
plt.ylabel("Total Global Market Share")
plt.legend(title="HS Code", labels=["Chemical Industrial Products (6)", "Mechanical/Electronics (16)", "Vehicle/Aircraft/Ship..etc (17)"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
